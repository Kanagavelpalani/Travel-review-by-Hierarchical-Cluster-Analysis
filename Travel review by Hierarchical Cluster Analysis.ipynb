7. Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.preprocessing import scale as s
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
data_final.shape
```

```python
z= sch.linkage(data_final, method = 'ward')
```

```python
z
```

```python
z.shape
```

```python
#create an dplt a Dendogram
den = sch.dendrogram(z)
plt.tick_params(
    axis = 'x',
    which = 'both',
    bottom = False,
    top =False,
    labelbottom = False)
plt.title("Hierarchical Cluster")
```

```python
den = sch.dendrogram(z)
plt.tick_params(
    axis = 'x',
    which = 'minor',
    bottom = False,
    top =False,
    labelbottom = False)
plt.title("Hierarchical Cluster")
```

```python
den = sch.dendrogram(z)
plt.tick_params(
    axis = 'x',
    which = 'major',
    bottom = False,
    top =False,
    labelbottom = False)
plt.title("Hierarchical Cluster")
```
